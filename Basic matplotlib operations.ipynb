```python
import numpy as np
import pandas as pd                     #pltoting matplotlib liberary 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a = np.arange(10) # assuming x value
a
```

```python
plt.plot(a, a**2, "rs", label="@2.2")          # ploting lines
plt.plot(a, a**2, "b--", label="^2.2")

plt.show()
```

```python
plt.plot(a, a**2, "rs", label="@2.2")
plt.plot(a, a**2, "b--", label="^2.2")           #showing legend
plt.legend()
plt.show()
```

```python
x = np.arange(10)
y = np.arange(10)
plt.plot(x,y, linewidth = 5.0)         #using linewidth
plt.show()
```

```python
x1 = [1,2,3,4,5,6]
y1 = [2,3,4,5,6,7]
x2 = [2,3,4,5,6,7]
y2 = [4,5,6,7,8,9]                             #ploting multiple lines
line = plt.plot(x1,y1,x2,y2)
plt.setp(line[0], color = "r",linewidth = 2.0)
plt.setp(line[1], color = "b",linewidth = 3.0)
plt.show()
```
